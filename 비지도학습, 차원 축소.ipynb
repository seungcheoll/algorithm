```python
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=300, noise=0.05, random_state=0)
```

```python
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c='gray',edgecolor='black',marker='o')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbs=DBSCAN(eps=0.2)
dbs.fit(X)
label_dbs=dbs.labels_
print(label_dbs)
```

```python
import pandas as pd
dbs_columns=['dbs_comp1','dbs_comp2']
X_dbs_df=pd.DataFrame(X,columns=dbs_columns)
X_dbs_df['target']=y
X_dbs_df['label_dbs']=label_dbs
X_dbs_df.head(5)
```

```python
print(set(X_dbs_df['target']))
print(set(X_dbs_df['label_dbs']))
```

```python
df=X_dbs_df
markers=['o','x']
for i,mark in enumerate(markers):
    df_i=df[df['label_dbs']==i]
    target_i=i
    X1=df_i['dbs_comp1']
    X2=df_i['dbs_comp2']
    plt.scatter(X1,X2, marker=mark, label=target_i)
    
plt.xlabel('dbs_component1')
plt.ylabel('dbs_component2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score=silhouette_score(X, label_dbs)
print(sil_score)
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=200, n_features=2, centers=2, random_state=2)
```

```python
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c='gray',edgecolor='black',marker='o')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(n_components=2, random_state=0)
gmm.fit(X)
label_gmm=gmm.predict(X)
print(label_gmm)
```

```python
import pandas as pd
gmm_columns=['gmm_comp1','gmm_comp2']
X_gmm_df=pd.DataFrame(X,columns=gmm_columns)
X_gmm_df['target']=y
X_gmm_df['label_gmm']=label_gmm
X_gmm_df.head(5)
```

```python
print(set(X_gmm_df['target']))
print(set(X_gmm_df['label_gmm']))
```

```python
df=X_gmm_df
markers=['o','x']
for i, mark in enumerate(markers):
    df_i=df[df['label_gmm']==i]
    target_i=i
    X1=df_i['gmm_comp1']
    X2=df_i['gmm_comp2']
    plt.scatter(X1,X2,marker=mark, label=target_i)
plt.xlabel('gmm_component1')
plt.ylabel('gmm_component2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score=silhouette_score(X,label_gmm)
print(sil_score)
```

```python
import pandas as pd
df=pd.read_csv('c:/data/data/wine_data.csv')
features=['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue',
       'Dilution', 'Proline']
X=df[features]
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_tn,X_te, y_tn, y_te=train_test_split(X,y,random_state=0)
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
st.fit(X_tn)
X_tn_st=st.transform(X_tn)
X_te_st=st.transform(X_te)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis()
lda.fit(X_tn_st, y_tn)
X_tn_lda=lda.transform(X_tn_st)
X_te_lda=lda.transform(X_te_st)
```
